```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("ML toolkit is Ready!!")
```

```python
X  = [[500] , [1000], [1500] ] #over here we have a 2D array, because scikit-learn expects inputs in the form of 2d array 
y = [100 , 200 , 300]

model = LinearRegression()
model.fit(X,y)
```

```python
prediction = model.predict([[1200]])
print(f"Predicted price for the 1200sq ft: ${prediction[0]*1000}")
```

```python
X_flat = np.array(X).flatten() #flatten the input for plotting

#plot
plt.scatter(X_flat,y, color = 'blue', label = 'Training Data')
plt.plot(X_flat, model.predict(X), color = 'red', label = 'Prediction Line')
plt.scatter([1200],prediction, color = 'green',label = 'Predicted Point')
plt.xlabel("Size(in sqft)")
plt.ylabel("Price in $1000s")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
